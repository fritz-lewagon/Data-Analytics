```python
from urllib.request import Request, urlopen
```

```python
url = 'https://a16z.com/portfolio/#crypto'
```

```python
def get_page_403(url):
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req, timeout=100).read()
    soup = bs4.BeautifulSoup(webpage, 'html.parser')
    return soup
```

```python
soup = get_page_403(url)
```

```python
all_codetags = soup.find_all(class_='portfolio__list--all')
first_listing = all_codetags[0]
```

```python
def get_url(tag_object):
    try:
        complete_object = tag_object.find(class_ = 'company__thumbnail company__thumbnail-link')
        url_part = complete_object.find("a")
        return url_part["href"]
    except:
        print("error")
```

```python
get_url(first_listing)
```

```python
def clean_url(url):
    if url.startswith("http://"):
        url = url[7:]
    if url.startswith("https://"):
        url = url[8:]
    if url.startswith("www."):
        url = url[4:]
    if url.endswith(".com"):
        url = url[:-4]
    return url
```

```python
clean_url(get_url(first_listing))
```

```python
all_ventures = []
for tag in all_codetags:
    all_ventures.append(clean_url(get_url(tag)))
```
